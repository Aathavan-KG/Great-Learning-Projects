### 11. Using a user defined function convert the ‘emp_len’ column into categorical column :

```python
pr2['emp_length'].unique()
```

```python
def exp(x):
    if x in ['< 1 year','1 year']:
        return 'fresher'
    elif x in ['2 years','3 years']:
        return 'Junior'
    elif x in ['4 years', '5 years', '6 years']:
        return 'Senior'
    elif x in ['7 years','8 years','9 years','10+ years']:
        return 'Expert'
    else:
        return x
```

```python
pr2['emp_length'].apply(exp)
```
